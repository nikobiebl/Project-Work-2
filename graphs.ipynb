## Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Results Classifier (random_state=42)

```python
df1 = pd.read_csv('results_classifier.csv') 

print(df1)
```

```python
ax = df1.set_index('Classifier').plot(kind='bar', figsize=(12, 8))

plt.title('Performance Comparison of Classifiers')
plt.xlabel('Classifier')
plt.ylabel('Score')
plt.ylim(0.92, 1.0)

plt.legend(title='Metrics')
plt.xticks(rotation=0)

ax.xaxis.grid(False)
ax.yaxis.set_major_locator(plt.MultipleLocator(0.005))

plt.show()
```

## Treshold Evaluation similar to Nebula

```python
df2 = pd.read_csv('treshold_evaluation.csv') 

print(df2)
```

```python
plt.figure(figsize=(12, 8))

plt.plot(df2['threshold'], df2['accuracy'], label='Accuracy')
plt.plot(df2['threshold'], df2['precision'], label='Precision')
plt.plot(df2['threshold'], df2['recall'], label='Recall')
plt.plot(df2['threshold'], df2['f1'], label='F1-Score')
plt.plot(df2['threshold'], df2['classifiable'], label='%-Classifiable')

plt.xlabel('Threshold')
plt.ylabel('Percentage')
plt.title('Impact of Threshold on Performance of the model and %-Classifiable')
plt.legend()
plt.grid(True)
plt.show()
```
